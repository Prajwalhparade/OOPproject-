```python
import nlpcloud

class NLPApp:
    def __init__(self):
        self.__database = {}
        self.__first_menu()

    def __first_menu(self):
        while True:
            first_input = input("""

            Hi! How would you like to proceed?

            1. Not a member? Register
            2. Already Member? Login
            3. Exit

            Enter your choice: """)

            if first_input == "1":
                self.__register()
                break

            elif first_input == "2":
                self.__login()
                break

            elif first_input == "3":
                print("Exit")
                exit()

            else:
                print("Invalid Input! Enter 1/2/3")

    def __register(self):
        username = input("Create Username: ")
        password = input("Create Password: ")

        if username in self.__database:
            print("User already exists! Try Login.")
            self.__first_menu()
        else:
            self.__database[username] = password
            print("Registration successful! Login now\n")
            self.__first_menu()

    def __login(self):
        username = input("Enter Username: ")
        password = input("Enter Password: ")

        if username in self.__database and self.__database[username] == password:
            print("Login Successful!\n")
            self.__second_menu()
        else:
            print("Invalid Username or Password!\n")
            self.__first_menu()

    def __second_menu(self):
        while True:
            user_input = input("""

            What would you like to do?

            1. Named Entity Recognition (NER)
            2. Language Detection
            3. Grammar & Spelling Correction
            4. Intent Classification
            5. Chatbot
            6. Sentiment Analysis
            7. Logout

            Enter your choice: """)

            if user_input == "1":
                self.__ner()

            elif user_input == "2":
                self.__language()

            elif user_input == "3":
                self.__gs_correction()

            elif user_input == "4":
                self.__intent_classification()

            elif user_input == "5":
                self.__chatbot()

            elif user_input == "6":
                self.__sentiment()

            elif user_input == "7":
                print("Logged out successfully!\n")
                self.__first_menu()
                break

            else:
                print("Invalid Input! Enter correct option.")



    def __ner(self):
        print("\n Named Entity Recognition ")
        text = input("Enter text: ")
        entity_type = input("Enter entity type : ")

        client = nlpcloud.Client("gpt-oss-120b","2e8d079c0cfd3af67f741966084b7c805ade823b",gpu=True)

        response = client.entities(text, searched_entity=entity_type)
        print("\nNER Result:", response)

    def __language(self):
        print("\n Language Detection ")
        text = input("Enter text: ")

        client = nlpcloud.Client("python-langdetect","2e8d079c0cfd3af67f741966084b7c805ade823b", gpu=False)

        response = client.langdetection(text)
        print("\nDetected Language:", response)

    def __gs_correction(self):
        print("\n Grammar & Spelling Correction ")
        text = input("Enter text to correct: ")

        client = nlpcloud.Client("gpt-oss-120b", "2e8d079c0cfd3af67f741966084b7c805ade823b", gpu=True)

        response = client.gs_correction(text)
        print("\nCorrected Sentence:", response)

    def __intent_classification(self):
        print("\n Intent Classification ")
        text = input("Enter text: ")

        client = nlpcloud.Client("gpt-oss-120b","2e8d079c0cfd3af67f741966084b7c805ade823b", gpu=True)

        response = client.intent_classification(text)
        print("\nIntent Detected:", response)

    def __chatbot(self):
        print("\n---- ChatBot (type 'exit' to stop) ----")

        client = nlpcloud.Client("gpt-oss-120b", "2e8d079c0cfd3af67f741966084b7c805ade823b", gpu=True)
        history = []

        while True:
            msg = input("You: ")
            if msg.lower() == "exit":
                print("Chat Ended!")
                break

            res = client.chatbot(msg, history=history)
            reply = res["response"]
            print("AI:", reply)
            history.append({"input": msg, "response": reply})

    def __sentiment(self):
        print("\n Sentiment Analysis  ")
        text = input("Enter text: ")

        client = nlpcloud.Client("gpt-oss-120b", "2e8d079c0cfd3af67f741966084b7c805ade823b", gpu=True)

        response = client.sentiment(text)
        print("\nSentiment Result:", response)


# -------- Start Application ----------
app = NLPApp()
```
